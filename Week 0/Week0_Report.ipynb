## 2. Mathematical Foundations

Key probability and statistics concepts were reviewed, including expectation, variance, and covariance. Expectation represents average outcomes, variance measures uncertainty or risk, and covariance describes how different assets move together. The normal distribution and z scores were introduced as standard tools for analyzing returns and deviations from the mean.

---

## 3. Geometric Brownian Motion

A discrete random walk models movement as random steps over time. As time steps become smaller, this process converges to Brownian motion, a continuous time stochastic process. Since Brownian motion can take negative values, it cannot model prices directly. By exponentiating Brownian motion, we obtain Geometric Brownian Motion (GBM), which ensures prices remain positive.

As part of the mini project, a GBM simulator was implemented using Python used to generate several possible future price paths from the same starting value. The resulting plots showed that prices can evolve very differently due to randomness alone. A histogram of final prices emphasized that outcomes should be understood as distributions rather than single predictions.

---

## 4. From Stochastic Motion to Asset Pricing

Financial markets are best modeled as stochastic systems. Asset pricing models do not aim to predict exact future prices but instead describe the range of possible outcomes and their probabilities. GBM provides a simple framework for linking stochastic motion to asset prices. This probabilistic approach is central to quantitative finance and underlies many pricing and risk models.

---

## 5. Monte Carlo Simulation and Option Valuation

Monte Carlo simulation is a key tool in quantitative finance. It works by generating many possible future scenarios and analyzing their outcomes. In option pricing, simulated price paths are used to compute payoffs, and the average payoff gives an estimate of the option's value. The core idea behind Monte Carlo methods are using randomness and repetition to estimate distributions of future outcomes.
